```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import os
```

# guitar

```python
data_dir = 'guitar/'
file_type = '.fig.mat'
files = os.listdir(data_dir)
files = [file for file in files if file.endswith(file_type)]

fig, ax = plt.subplots(len(files), 1, figsize=(8, 3 * len(files)))

for idx, file in enumerate(files):
    path = data_dir + file
    data = loadmat(path)
    freq = data['frequency'].flatten()
    acc = data['accelerance'].flatten()

    ax[idx].plot(freq, acc)
    ax[idx].set_ylabel('accelerance (dB)')
    ax[idx].set_xlim(50, 1000)
    ax[idx].set_title(file)

plt.tight_layout()
plt.show()

```
